```python
import matplotlib.pyplot as plt
```

## General Plotting
- Plot Types: bar, barbs, barh, boxplot, hist, pie, plot, plot_date, polar, psd, quiver, scatter, specgram, spy, stackplot, stairs, stem, step, streamplot, violin

```python
import numpy as np

# Sample Data
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

# Execution
fig, axs = plt.subplots(2, 2, figsize=(5, 4), sharex=True, sharey=False)

axs[0, 0].plot(x, y, label='one')
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].plot(x, y, 'tab:orange', label='two')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].plot(x, -y, 'tab:green', label='three')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].plot(x, -y, 'tab:red', label='four')
axs[1, 1].set_title('Axis [1, 1]')

for ax in axs.flat:
    # Option 1
    ax.set(xlabel='x-label', ylabel='y-label')
    
    # Option 2
    ax.set_xlabel('x-label')
    ax.set_ylabel('y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()

# Set legend
for ax in axs.flat:
    ax.legend(loc='lower right')

# Add text
for ax in axs.flat:
    ax.text(0.1, 0.2, 'text', transform=ax.transAxes)

# Horizontal and Vertical Lines
axs[0,0].axhline(y=1)

# Formatting
fig.suptitle('Vertically stacked subplots')
fig.tight_layout()

plt.show()
print()
```
