## Array Operations

- Performed element-wise

```python
np.sum(axis=0)                      # sums (axis=0: rows,   axis=1: columns; keepdims=True: maintains as multi-dim array)
np.min(axis=0)                      # minimum
np.max(axis=0)                      # maximum
np.mean(axis=0)                     # mean
np.argmax()                         # index of maximum value
np.nanargmax()                      # index of maximum value without na
np.argmin()                         # index of minimum value
np.nanargmin()                      # index of minimum value without na
np.argpartition(nd1, -4)[-4:]       # indices of four largest elements

np1.multiply(np2)                   # matrix multiplication
nd1 / nd2                           # applies to other operators 
                                    # (maintains highest type, e.g. float and int) 
nd1 * 2                             # multiplies element-wise
```
